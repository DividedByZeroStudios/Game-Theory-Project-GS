```python
def baseitemvalues(hero, round):
    """Returns the base values for the payoff matrix"""
    itemvaluelist = []
    if hero in {"PA", "pa", "Phantom Assassin"}:
        x = PAitems
    elif hero in {"slark","Slark"}:
        x = Slarkitems
    else:
        x = "Hero not yet considered"
    
    for item in dotaitems:
        
        itemPage = requests.get("https://dota2.gamepedia.com/" + item)
        tempsoup = BeautifulSoup(itemPage.text, 'html.parser')
        rawtext2 = tempsoup.get_text()
        webpagewords = rawtext2.split()


        for word in webpagewords:
            if word.find('Cost') > -1:
                itemcost = int(word.strip("Cost")) #this is the cost of the item
                break
        if round == 1:        
            if itemcost < 2000:
                morepoints = 2 #gives the players more incentive to buy cheaper items earlier on, as it makes more sense
            else:
                morepoints = 0 
        elif round < 5:
            if 2000 < itemcost < 4500:
                morepoints = 2 #gives the players more incentive to buy cheaper items earlier on, as it makes more sense
            else:
                morepoints = 0 
        elif round > 4:
            morepoints = 0
        
        if item in x:
            itemvaluelist.append(2 + morepoints)
            
        else:
            itemvaluelist.append(0 + morepoints)
    
    print(itemvaluelist)
    
```
